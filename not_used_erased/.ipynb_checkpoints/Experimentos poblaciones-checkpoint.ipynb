```python
import random
import time
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx
import math
from collections import Counter
import os
import sys

# Notebook path
current_directory = os.getcwd()
# Get one behind for calling the other modules
root_directory = os.path.abspath(os.path.join(current_directory, '..'))
sys.path.append(root_directory)

#Classes
from model_class.UD import UD
from model_class.Variety import Variety

#Network and measures
from to_get_data.net_construction_vis_measures import *
#Agrobiodiversity
from to_get_data.agrobiodiversity_data import *
```

```python
unidades_iniciais = 30
beta = 0.06
meia_filhos = 4
tiempo = 201
herencia = 0
punto_vista = 20
total_vivas = []
experimentos = 5
tempo = list(range(0,tiempo,punto_vista))
for i in range(experimentos):
    print(f"experimento {i}")
    UD.uds = {}
    instancias = [UD(id,meia_filhos)  for id in range(1, unidades_iniciais+1)]
    for id_ud, ud in UD.uds.items():
        for _ in range(5):
            ud.varieties.append(Variety())
    # Lists
    vivas = []
    mortas = []
    exp = 0
    for t in range(tiempo):
        uds_copy = dict(UD.uds)
        any_ud_active = False
        for id_ud, ud in uds_copy.items():
            if ud.activa == True:
                any_ud_active = True
                ud.ter_filho()
                ud.buscar_ud(meia_filhos, 0)
                ud.incrementar_idade()
                ud.death_probability(beta)
        if not any_ud_active:
            print("No hay mas")
            # Only 0
            vivas.extend([0] * (int((tiempo/punto_vista)-(len(vivas)))))
            #mortas.extend([unidades_iniciais] * (tiempo - t))
            break
        # Collect data
        if t % punto_vista == 0:
            viv = len([ud for id, ud in uds_copy.items() if ud.activa])
            mort = len([ud for id, ud in uds_copy.items() if not ud.activa])
            vivas.append(viv)
            mortas.append(mort)
        #if t%100 == 0:
            #print(f"tamos en tiempo {t} con vivas {viv}")
    total_vivas.append(vivas)
```

```python
# Mean and std
medias = np.mean(total_vivas, axis=0)
desviaciones_estandar = np.std(total_vivas, axis=0)

# graph
plt.errorbar(tempo, medias, yerr=desviaciones_estandar, fmt='o-', label='Meia')
plt.xlabel('Tempo (t)')
plt.ylabel('Quantidade de UDs')
plt.legend()
plt.show()
```
